```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
death_row = pd.read_csv(r"C:\Users\user\Desktop\Texas_Last_Statemen.csv", encoding='ISO-8859-1')
death_row
```

```python
death_row.isnull().mean() *100
```

```python
death_row = death_row.dropna(subset=["LastStatement"])
death_row
```

```python
for index, row in death_row.head(30).iterrows():
    sentence = str(row["LastStatement"])
    if "sorry" in sentence.lower() or "please" in sentence.lower():
        print(f"{row['LastName'].upper()}: {sentence}")
```

```python
death_row["CountyOfConviction"].value_counts()
```

```python
death_row["Age"].value_counts()
```

```python
avg_age = death_row.groupby("Race")["Age"].mean()
avg_age.plot(kind="bar", color="skyblue")
plt.title("Average Age per Race")
plt.ylabel("Average Age")
plt.xlabel("Race")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x="Race", y="Age", data=death_row)
plt.title("Age Distribution by Race")
plt.xticks(rotation=45)
plt.show()
```
